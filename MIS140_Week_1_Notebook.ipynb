## Row Filter
image.png
Filtering data allows you to focus on specific **customers**, **transactions**, or **events** that match certain criteria. Pandas makes it easy to filter rows using **conditional expressions** and **boolean logic**.


### Filtering Based on a Single Condition
To filter rows, place a **condition** inside square brackets `[]`.  

#### Example: Customers born before 1980  

```python
older_customers = df[df['Year_Birth'] < 1980]
```

- This selects all rows where **Year_Birth** is **less than 1980**.  
- The result is a **filtered DataFrame** with only those customers.  

You can check the size of the filtered dataset using:  

```python
older_customers.shape  # (Number of matching rows, columns)
```

üí° **Try modifying the condition** to select customers born after 1990.  

---

````python
older_customers = df[df['Year_Birth'] > 1990]
older_customers.shape  # (Number of matching rows, columns)
display(older_customers)
````

### Filtering Based on Multiple Conditions
Use **`&` (AND)** to filter rows that match **all** conditions.  
Use **`|` (OR)** to filter rows that match **at least one** condition.  

#### Example: Customers born before 1980 AND earning more than $70,000  

```python
older_high_income = df[(df["Year_Birth"] < 1980) & (df["Income"] > 70000)]
```

#### Example: Customers with Kids OR Teens at home  

```python
parents = df[(df["Kidhome"] > 0) | (df["Teenhome"] > 0)]
```

- This selects customers who have at least **one kid** OR **one teenager** at home.  

---

````python
parents = df[(df["Kidhome"] >= 1) | (df["Teenhome"] >= 1)]
display(parents)
````

### Filtering for Specific Categories

To filter based on **categorical values**, use `.isin()`.  

#### Example: Customers with "Graduation" or "PhD" education levels  

```python
educated_customers = df[df["Education"].isin(["Graduation", "PhD"])]
```

````python
educated_customers = df[df["Education"].isin(["Graduation", "PhD"])]
display(educated_customers)
````

## Specific Rows and Columns Selection
image.png

When analyzing data, you may need to extract specific **rows** and **columns** based on conditions or positions. Pandas provides two powerful methods for this:  

- **`.loc[]`** ‚Äì Selects rows and columns using **labels** (column names or row index values).  
- **`.iloc[]`** ‚Äì Selects rows and columns using **positions** (row and column index numbers).  

### Selecting Rows and Columns Using `.loc[]` (Label-Based Selection)

Use `.loc[]` when selecting specific **rows and columns by name**.

#### Example: Select "Education" for customers born before 1980  

```python
education_old_customers = df.loc[df["Year_Birth"] < 1980, "Education"]
```

- The **first part** (`df["Year_Birth"] < 1980`) filters rows where `Year_Birth` is **less than 1980**.  
- The **second part** (`"Education"`) selects only the **Education** column.  

#### Example: Select "Education" and "Income" for customers born before 1980  

```python
old_customers_info = df.loc[df["Year_Birth"] < 1980, ["Education", "Income"]]
```

üí° **Try modifying this to select "Marital_Status" instead.**  

````python
df.loc[df["Year_Birth"] < 1980, "Marital_Status"]
````

### Selecting Rows and Columns Using `.iloc[]` (Position-Based Selection)

Use `.iloc[]` when selecting **rows and columns by their numerical index position**.

#### **Example: Select rows 10 to 25 and columns 3 to 5**  

```python
subset_data = df.iloc[9:25, 2:5]
```

- **Rows 10 to 25** (Python indexing starts at **0**, so row 10 is at index **9**).  
- **Columns 3 to 5** (index **2 to 4**, since the last number is **exclusive**).  

üí° **Modify this to select the first 15 rows and columns 1 to 3.**  


````python
subset_data = df.iloc[0:15, 1:3]
display(subset_data)
````

### Updating Values in a DataFrame
You can also update specific values in a DataFrame using `.iloc[]`.

#### Example: Set "Education" to "Unknown" for the first 3 customers  

```python
df.iloc[0:3, df.columns.get_loc("Education")] = "Unknown"
```

üí° **Try changing "Marital_Status" to "Not Disclosed" for the first 5 rows.**  

---

````python
df.iloc[0:5, df.columns.get_loc("Marital_Status")] = "Not Disclosed"
display(df["Marital_Status"])
````
